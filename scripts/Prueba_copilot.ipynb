```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
sales_data = {
    'department': ['Sales', 'Sales', 'Marketing', 'Marketing', 'Operations', 'Operations'],
    'product': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Monitor', 'USB Cable'],
    'quantity': [5, 15, 8, 12, 10, 25],
    'unit_price': [800, 25, 800, 60, 300, 10]
}

# Añadir columnas de precio sin IVA y con IVA
sales_data['precio_sin_iva'] = sales_data['unit_price']
sales_data['precio_con_iva'] = [round(p * (1 + IVA), 2) for p in sales_data['unit_price']]
IVA = 0.21
del sales_data['unit_price']

df_ventas = pd.DataFrame(sales_data)

# Mostrar el DataFrame resultante
print(df_ventas)
```

```python
# Resumen estadístico del DataFrame df_ventas
df_ventas.describe(include='all')
```

```python
# Gráfico de barras: ventas totales por departamento
import matplotlib.pyplot as plt

ventas_por_departamento = df_ventas.groupby('department')['precio_con_iva'].sum()
ventas_por_departamento.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Departamento')
plt.ylabel('Ventas Totales')
plt.title('Ventas Totales por Departamento')
plt.tight_layout()
plt.show()
```
